## Crimean Tatar to English/Russian/Ukrainian parallel corpora 

```python
import pandas as pd

splits = {"crh_eng": "eng_crh.csv", "crh_rus": "rus_crh.csv", "crh_ukr": "ukr_crh.csv"}
```

### latn

```python
df = pd.read_csv("hf://datasets/QIRIM/crh-parallel-corpora/" + splits["crh_eng"])
df = df[(df["crh_Latn"].notna()) & (df["crh_Cyrl"].isna())][["crh_Latn", "target_lang"]]
df.to_csv("data/parallel_corpora/crh_Latn_eng.csv", index=False)
df.size
```
